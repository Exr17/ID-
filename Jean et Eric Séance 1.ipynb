<div style="background-color:tomato;border-style:solid;border-width:1px;padding-left:40px;padding-right:40px;line-height:1.2;text-align:center;font: 20px Courier New;">
<br />Exercice : que fait ce code ? Quel est la fonction de timeit ?<br />
<br />
</div>

````python
import random
import timeit
TAX_RATE = .08
txns = [random.randrange(100) for _ in range(100000)]
def get_price(txn):
    return txn * (1 + TAX_RATE)

def get_prices_with_map():
    return list(map(get_price, txns))

def get_prices_with_comprehension():
    return [get_price(txn) for txn in txns]

def get_prices_with_loop():
    prices = []
    for txn in txns:
        prices.append(get_price(txn))
    return prices

tmap = timeit.timeit(get_prices_with_map, number=100)
tListComp = timeit.timeit(get_prices_with_comprehension, number=100)
tLoop = timeit.timeit(get_prices_with_loop, number=100)

print("timeit.timeit(get_prices_with_map, number=100) = {}".format(tmap))
print("timeit.timeit(get_prices_with_comprehension, number=100) = {}".format(tListComp))
print("timeit.timeit(get_prices_with_loop, number=100) = {}".format(tLoop))
````

Le code précédent permet de calculer le prix TTC à partir des prix HT de 3 manières différentes : en utilisant la fonction map, par compréhension ou en utilisant une boucle.
La fonction timeit permet de calculer la durée de l'éxécution de la fonction un nombre number fois (ici 100)

<br />Création d'un graphe avec networkx (une façon parmi d'autres)<br />
<br />

````python
import networkx as nx
import matplotlib.pyplot as plt


def add_edge_to_graph(G, e1, e2, w):
    G.add_edge(e1, e2, weight=w)


G = nx.Graph()
points = [(1, 10), (8, 10), (10, 8), (7, 4), (3, 1)]  # (x,y) points
edges = [(0, 1, 10), (1, 2, 5), (2, 3, 25), (0, 3, 3), (3, 4, 8)]  # (v1,v2, weight)

for i in range(len(edges)):
    add_edge_to_graph(G, points[edges[i][0]], points[edges[i][1]], edges[i][2])

# you want your own layout
# pos = nx.spring_layout(G)
pos = {point: point for point in points}

# add axis
fig, ax = plt.subplots()

nx.draw(G, pos=pos, node_color='r', node_size=1500, ax=ax)  # draw nodes and edges
nx.draw_networkx_labels(G, pos=pos)  # draw node labels/names
# draw edge weights
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, ax=ax)
plt.axis("on")
ax.set_xlim(0, 11)
ax.set_ylim(0,11)
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
plt.show()
````

<br />Création "manuelle" d'un graphe de besoins : les sommets, leurs positions, les arcs et leurs valuations (comportant des unités) sont fournis. Un calcul de besoins est ensuite effectué. Examiner soigneusement le code.<br /><br />
</div>

````python
import networkx as nx
import matplotlib.pyplot as plt

# définition d'un fonction extrayant la partie numérique d'une chaine
# de la forme '0.001h'
# exemple : ValArcBesoin('0.001h')= 0.001
def ValArcBesoin(S):
    if len(S)==1:return float(S)
    if S[-1].isalpha() ==  False:return float(S)
    return float(S[:len(S)-1])


"""
Couleurs MatPlotLib
b : blue.
g : green.
r : red.
c : cyan.
m : magenta.
y : yellow.
k : black.
w : white.
"""

# initialiser le graphe
g = nx.DiGraph()

# donner la liste des sommets et leurs positions
LSom = {'Boite' : (0,1.5), 
        'Fond' : (1,3), 
        'Couvercle' : (1,0), 
        'Melange' : (2.3,0), 
        'MO' : (3.5,3), 
        'PVC' : (3.5,1.5),
        'Colorant' : (3.5,0)}

# créer un vecteur des couleurs des sommets
color_nodes = ['gray']*len(LSom)


# donner la liste des arcs et leurs poids
LEdges = {('Boite','Fond'):'1',
          ('Boite','Couvercle'):'1',
          ('Boite','MO'):'0.001h',
          ('Boite','Melange'):'1g',
          ('Fond','MO'):'0.002h',
          ('Fond','Melange'):'20g',
          ('Couvercle','MO'):'0.001h',
          ('Couvercle','Melange'):'10g',
          ('Melange','MO'):'0.000001h',
          ('Melange','PVC'):'1g',
          ('Melange','Colorant'):'0.001g'
        }

# ajouter les sommets et leurs positions au graphe
for v in LSom:
    g.add_node(v, pos = LSom[v])
    
# ajouter les arcs et leurs poids au graphe
for e in LEdges:
    g.add_edge(*e,weight=LEdges[e])

# obtenir le dictionnaire des positions des sommets (inutile ici car il s'agit de LSom)
pos=nx.get_node_attributes(g,'pos')
    
# calcul des besoins selon l'initialisation P(Boite)=1000 et le calcul
# P(x) = somme sur les antécédents y de x, de P(y)*Conso(y,x)
Besoins = {}
Besoins['Boite'] = 1000
for v in nx.topological_sort(g):
    if v != 'Boite':
        Besoins[v] = 0
        for p in g.predecessors(v):
            #print "v=",v,", p=",p
            Qt = ValArcBesoin(LEdges[(p, v)])
            Besoins[v]+=Besoins[p]*Qt

# définition des unités attachées à chaque sommet
Unites = {'Boite':'', 'Fond':'', 'Couvercle':'', 'Melange':'g', 'MO':'h', 'PVC':'g', 'Colorant':'g'}


fig, ax = plt.subplots(figsize=(12, 8))

#plt.figure(figsize=(8, 6))
nx.draw(g,with_labels=True, arrows=True, pos = pos, node_size=3000,node_color = color_nodes)

"""
nx.draw_networkx_edge_labels(g, pos = pos, arrowsize=30,
                             edge_labels= LEdges,
                             font_size=14 )
"""
nx.draw_networkx_edge_labels(g, pos = pos, 
                             edge_labels= LEdges,
                             font_size=14 )

for s in LSom:
    ax.text(LSom[s][0]-0.2, LSom[s][1]+0.3, "{}{}".format(Besoins[s], Unites[s]), color='red', fontweight='bold', fontsize=18 )
    
    
#plt.draw()
plt.axis('off')
plt.show()
````

````python
# D'après https://github.com/MUSoC/Visualization-of-popular-algorithms-in-Python/blob/master/Graph%20Coloring/graph_coloring.py

import networkx as nx
import matplotlib.pyplot as plt


#implementation of welsh_powell algorithm
def welsh_powell(G):
    #sorting the nodes based on it's valency
    #node_list = sorted(G.nodes(), key =lambda x:G.neighbors(x))
    LN = [(v, len([n for n in G.neighbors(v)])) for v in G.nodes]
    LNS = sorted(LN, key=lambda x: x[1], reverse=True)
    node_list = [v[0] for v in LNS]
    
    col_val = {} #dictionary to store the colors assigned to each node
    col_val[node_list[0]] = 0 #assign the first color to the first node
    # Assign colors to remaining N-1 nodes
    for node in node_list[1:]:
        available = [True] * len(G.nodes()) #boolean list[i] contains false if the node color 'i' is not available

        #iterates through all the adjacent nodes and marks it's color as unavailable, if it's color has been set already
        for adj_node in G.neighbors(node): 
            if adj_node in col_val.keys():
                col = col_val[adj_node]
                available[col] = False
        clr = 0
        for clr in range(len(available)):
            if available[clr] == True:
                break
        col_val[node] = clr
    print(col_val)
    return col_val


#takes input from the file and creates a undirected graph 
def CreateGraph():
    G = nx.Graph()
    f = open('GraphColor.txt')
    #f = open('SessionExamen.txt', encoding='utf-8')
    n = int(f.readline())
    for i in range(n):
        graph_edge_list = f.readline().split()
        G.add_edge(graph_edge_list[0], graph_edge_list[1]) 
    return G


#draws the graph and displays the weights on the edges
def DrawGraph(G,col_val):
    pos = nx.spring_layout(G)
    values = [col_val.get(node, 'blue') for node in G.nodes()]
    LColor = ['red', 'cyan', 'orange', 'gray', 'green']
    ColorV = [LColor[v] for v in values]
    nx.draw(G, pos, with_labels = True, node_color = ColorV, edge_color = 'black' ,width = 1, alpha = 1, node_size=4000)  #with_labels=true is to show the node number in the output graph
    #nx.draw(G, pos, with_labels = True, node_color = color_nodes, edge_color = 'black' ,width = 1, alpha = 1)



#main function
if __name__ == "__main__":
    plt.figure(figsize=(8, 8))
    G = CreateGraph()
    col_val = welsh_powell(G)
    DrawGraph(G,col_val)
    #plt.show()
````

**Principe de l’heuristique de Welsh-Powell** :
On commence par examiner les nœuds du graphe en partant de celui qui a le plus de voisins (degré le plus élevé) jusqu’à celui qui en a le moins.
Le premier nœud est coloré avec une couleur quelconque. Pour chaque nœud suivant, on regarde quelles couleurs ont déjà été attribuées à ses voisins, puis on choisit une couleur qui n’a pas encore été utilisée dans par ses voisins
En colorant d’abord les nœuds les plus connectés, on s’occupe en priorité des cas les plus contraignants, ce qui facilite l’attribution des couleurs aux autres nœuds et limite le nombre total de couleurs utilisées.
